```python
df['Heart Disease'] = df['Heart Disease'].map({'Presence':1,'Absence':0})
```

```python
df.head()
```

```python
X=df.drop(['id','Heart Disease'],axis=1)
y=df['Heart Disease']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val = train_test_split(X,y,stratify=y,test_size=0.2,random_state=42)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
encoder = OneHotEncoder()

categorical_colms_train = ['Chest pain type','EKG results','Slope of ST','Number of vessels fluro',
                            'Thallium']
num_columns = ['Age','BP','Cholesterol','Max HR','ST depression']

ct = ColumnTransformer([('encoder',encoder,categorical_colms_train),
                       ('scaler',scaler,num_columns)],remainder = 'passthrough')
X_train_transformed = ct.fit_transform(X_train)
X_val_transformed = ct.transform(X_val)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
log = LogisticRegression()
log.fit(X_train_transformed,y_train)
log_pred = log.predict(X_val_transformed)
```

```python
log_train_pred = log.predict(X_train_transformed)
print(accuracy_score(log_train_pred
                     ,y_train))
```

```python
print(accuracy_score(log_pred
                     ,y_val))
```

```python
print("Train Report:\n", classification_report(y_train, log_train_pred))
print("Validation Report:\n", classification_report(y_val, log_pred))
```

```python
print("Validation Confusion Matrix:\n", confusion_matrix(y_val, log_pred))
```

```python
from sklearn.metrics import roc_auc_score

y_train_proba = log.predict_proba(X_train_transformed)[:,1]
y_val_proba = log.predict_proba(X_val_transformed)[:,1]

print("Train ROC-AUC:", roc_auc_score(y_train, y_train_proba))
print("Validation ROC-AUC:", roc_auc_score(y_val, y_val_proba))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train_transformed,y_train)
train_pred = clf.predict(X_train_transformed)
val_pred = clf.predict(X_val_transformed)
```

```python
print("Train Accuracy:", accuracy_score(y_train, train_pred))
print("Validation Accuracy:", accuracy_score(y_val, val_pred))
```

```python
print("Train Report:\n", classification_report(y_train, train_pred))
print("Validation Report:\n", classification_report(y_val, val_pred))
```

```python
clf_train_proba = clf.predict_proba(X_train_transformed)[:,1]
clf_val_proba = clf.predict_proba(X_val_transformed)[:,1]

print("Train ROC-AUC:", roc_auc_score(y_train, clf_train_proba))
print("Validation ROC-AUC:", roc_auc_score(y_val, clf_val_proba))
```
